```python
import pandas as pd
import numpy as np
pd.options.display.max_columns=None
pd.options.display.max_rows=None

import scipy.stats as stats
```

```python
df = pd.read_csv('sample_train_good_40k.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
dtypes = df.apply(lambda col: col.dtype)
null_count = df.isnull().sum()
nuniques = df.apply(lambda col:col.nunique())

df_info = pd.DataFrame({'dtypes':dtypes,'null_count':null_count,'nuniques':nuniques})
df_info
```

```python
num_cat_col = df_info[(df_info['dtypes']!='object')&(df_info['nuniques']<4)].index
num_col = df_info[(df_info['dtypes']!='object')&(df_info['nuniques']>=4)].index
cat_col = df_info[df_info['dtypes']=='object'].index
```

```python
cat_col = list(num_cat_col) + list(cat_col)
```

```python
target = 'TARGET'
```

```python
cat_col.remove('TARGET')
```

## Numerical vs Target(cat)

```python
num_col
```

```python
l = []
alpha=0.05
for col in num_col:
    not_paid = df[df[target] == 1][col].dropna()
    paid = df[df[target] == 0][col].dropna()
    t, p = stats.f_oneway(paid,not_paid)
    #print(col, 'p-value:', p)
    if p>alpha/2:
        l.append((col,1))
    else:
        l.append((col,0))
```

```python
sig_num_col = [i[0] for i in l if i[1]==0]# ==0 checks for siginificant columns
print(sig_num_col)
```

```python
len(sig_num_col)
```

### Categorical vs Target(cat)

```python
print(cat_col)
```

```python
l = []
for col in cat_col:
    contingency_table = pd.crosstab(df[col], df[target])
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    print(col, 'p-value:', p)
    if p>alpha/2:
        l.append((col,1))
    else:
        l.append((col,0))
```

```python
sig_cat_col = [i[0] for i in l if i[1]==0]
print(sig_cat_col)
```
